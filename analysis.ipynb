# Importaçao de bibliotecas

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

# Carregamento dos dados

```python
df = pd.read_csv("sales_data.csv")
df.head()
```

# Criaçao das metrica de (faturamento) PYTHON 
# anali-se das colunas descritas 

```python
print(df.columns)
```

```python
# Criar a métrica de faturamento
df["faturamento"] = df["Quantity Ordered"] * df["Price Each"]

# Visualizar os dados
df.head()
```

# Analise __Faturamento por produto 

```python
import matplotlib.pyplot as plt
import numpy as np

# Agrupar dados
dados = df.groupby("Product")["faturamento"].sum().sort_values()

# Criar cores em gradiente
cores = plt.cm.viridis(np.linspace(0.2, 0.8, len(dados)))

# Criar gráfico
plt.figure(figsize=(10, 6))
bars = plt.barh(dados.index, dados.values, color=cores)

# Adicionar valores nas barras
for i, v in enumerate(dados.values):
    plt.text(v, i, f" R$ {v:,.0f}", va='center')

# Títulos e rótulos
plt.title("Faturamento por Produto", fontsize=16, weight='bold')
plt.xlabel("Faturamento (R$)", fontsize=12)
plt.ylabel("Produto", fontsize=12)

# Melhorar layout
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()
```

# CÓDIGO COMPLETO (DASHBOARD)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# =========================
# CARREGAR DADOS
# =========================
df = pd.read_csv("sales_data.csv")  # <-- CORRIGIDO (sem .csv duplicado)

# =========================
# TRATAMENTO DOS DADOS
# =========================

# Corrigir data (SEM WARNING)
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    format="%m/%d/%y %H:%M",
    errors="coerce"
)

# Criar mês e nome do mês
df["Month"] = df["Order Date"].dt.month
df["Month_Name"] = df["Order Date"].dt.month_name()

# Criar faturamento
df["faturamento"] = df["Quantity Ordered"] * df["Price Each"]

# =========================
# KPIs
# =========================

faturamento_total = df["faturamento"].sum()

produto_agg = df.groupby("Product")["faturamento"].sum()
produto_top = produto_agg.idxmax()
valor_produto_top = produto_agg.max()

# Ordenar meses corretamente
ordem_meses = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

mes_agg = df.groupby("Month_Name")["faturamento"].sum().reindex(ordem_meses)
mes_top = mes_agg.idxmax()

cidade_agg = df.groupby("City")["faturamento"].sum()
cidade_top = cidade_agg.idxmax()

# =========================
# DASHBOARD
# =========================

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# -------- Produto --------
dados_produto = produto_agg.sort_values()
cores = plt.cm.viridis(np.linspace(0.2, 0.8, len(dados_produto)))

bars = axes[0,0].barh(dados_produto.index, dados_produto.values, color=cores)

for bar in bars:
    axes[0,0].text(
        bar.get_width(),
        bar.get_y() + bar.get_height()/2,
        f"{bar.get_width():,.0f}",
        va='center'
    )

axes[0,0].set_title("Faturamento por Produto")

# -------- Mês --------
bars_mes = axes[0,1].bar(mes_agg.index, mes_agg.values, color="skyblue")

for bar in bars_mes:
    axes[0,1].text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f"{bar.get_height():,.0f}",
        ha='center',
        va='bottom',
        fontsize=8
    )

axes[0,1].set_title("Faturamento por Mês")
axes[0,1].tick_params(axis='x', rotation=45)

# -------- Cidade --------
dados_cidade = cidade_agg.sort_values()

bars_city = axes[1,0].barh(dados_cidade.index, dados_cidade.values, color="lightgreen")

for bar in bars_city:
    axes[1,0].text(
        bar.get_width(),
        bar.get_y() + bar.get_height()/2,
        f"{bar.get_width():,.0f}",
        va='center'
    )

axes[1,0].set_title("Faturamento por Cidade")

# -------- KPIs --------
axes[1,1].axis('off')

texto = f"""
 FATURAMENTO TOTAL
R$ {faturamento_total:,.2f}

 PRODUTO TOP
{produto_top}
R$ {valor_produto_top:,.2f}

 MELHOR MÊS
{mes_top}

 MELHOR CIDADE
{cidade_top}
"""

axes[1,1].text(0, 0.5, texto, fontsize=12)

# Layout final
plt.tight_layout()
plt.show()
```
